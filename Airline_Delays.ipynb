FILTRADO

```python
#RETRASOS A CAUSA DEL CLIMA

for d in collection.find(
    {"weather_delay": {"$gte": 60,
            "$lte": 90},
    "arr_delay": {"$gt": 10}
    },
    {
        "_id": 0,
        "weather_delay": 1,
        "airport" : 1,
        "arr_delay" : 1, 
    }
).sort("weather_delay", -1).limit(20):
    print(d)
```

ACTUALIZACIONES

```python

result = collection.update_many(
    {
        "year": 2019,
        "month": 9,
        "arr_flights": {"$gt": 100}
    },
    [
        {
            "$set": {
                "busy_airport": True,
                "delay_percentage": {
                    "$cond": [
                        {"$gt": ["$arr_flights", 0]},
                        {
                            "$multiply": [
                                {"$divide": ["$arr_del15", "$arr_flights"]},
                                100
                            ]
                        },
                        0
                    ]
                },
                "updated_at": datetime.utcnow()
            }
        }
    ]
)

print("Matched:", result.matched_count)
print("Modified:", result.modified_count)
```

AGREGACIONES

```python
#TOTAL DE RETRASOS POR AERONAVE

pipeline = [
    {"$group": {
        "_id": "$carrier_name",
        "total_delay": {"$sum": "$late_aircraft_delay"}
    }},
    {"$sort": {"total_delay": -1}},
    {"$limit": 6}
]

for d in collection.aggregate(pipeline):
    print(d)
```

```python
#RETRASO MEDIO EN CADA AEROPUERTO
pipeline = [
    {
        "$group": {
            "_id": "$airport",
            "avg_delay": {"$avg": "$arr_delay"}
        }
    },
    {"$sort": {"avg_delay": -1}},
    {"$limit": 10}
]

for d in collection.aggregate(pipeline):
    print(d)
```

```python
#LOS 10 AEROPUERTOS CON MAS RETRASOS
pipeline = [
    {"$group": {
        "_id": "$airport_name",
        "total_delay": {"$sum": "$arr_delay"}
    }},
    {"$sort": {"total_delay": -1}},
    {"$limit": 10}
]

for d in collection.aggregate(pipeline):
    print(d)
```
